```python
import pandas as pd 
import plotly.express as px

data = pd.read_csv("Netflix Userbase.csv")
data.info()
data.head()
```

```python
#Removing the data that will not add anything to the analysis.

data = data.drop(["User ID", "Join Date", "Last Payment Date"], axis=1)
data.info()
```

```python
# Visualization
# Gender Count

gender_count = px.histogram(data, x="Gender", color="Gender",
                            title="Gender Count", opacity=0.87)
gender_count.show()
```

```python
#Age Distribution

age = px.violin(data, x="Age",title="Age Distribution")
age.show()
```

```python
#Subscription Types

subscription = px.pie(data, data["Subscription Type"].value_counts().index, 
                      data["Subscription Type"].value_counts(normalize=True), 
                      title="Subscription Types", opacity=0.87)
subscription.show()
```

```python
#Device Distribution

devices_chart = px.histogram(data, x="Device", color="Device",
                             title="Device Distribution", opacity=0.87)
devices_chart.show()
```

```python
#Plan Duration

plan_duration = px.histogram(data, x="Plan Duration", color="Plan Duration",
                             title="Plan Duration" ,opacity=0.87)
plan_duration.show()

#It's quite simple but is interesting to know that all subscriptions are only 1 month.
#Imagine all the possibilities that could came from only this data.
```

```python
#Users by Country

country_chart = px.histogram(data, y="Country", color="Country",
                             title="Users by Country", opacity=0.87)
country_chart.show()
```

```python
#Monthly Revenue by Gender

monthly_revenue_by_gender = px.pie(data, data["Gender"].value_counts().index,
                                 data["Gender"].value_counts(normalize=True),
                                 title="Monthly Revenue by Gender", opacity=0.87)
monthly_revenue_by_gender.show()
```

```python
#Monthly Revenue by Subscription Type

monthly_revenue_by_subscription = px.pie(data, data["Subscription Type"],
                                        data["Monthly Revenue"], 
                                         title="Monthly Revenue by Subscription Type",
                                        opacity=0.87)
monthly_revenue_by_subscription.show()
```

```python
#Monthly Revenue by Country

monthly_revenue_by_country = px.pie(data, data["Country"], 
                                    data["Monthly Revenue"],
                                   title="Monthly Revenue by Country", opacity=0.87)
monthly_revenue_by_country.show()
```
